## 1. Why is NumPy better than lists?
NumPy is better than lists for several reasons:

Performance: NumPy arrays are more efficient for numerical computations as they are implemented in C and optimized for performance.

Memory: NumPy arrays are more memory-efficient as they store elements of the same type, unlike lists that can store elements of different types.

Convenience: NumPy provides a wide range of mathematical functions and operations that can be applied to arrays, making it easier to perform complex calculations.

Vectorization: NumPy supports vectorized operations, which are faster than using loops for element-wise operations on lists.

## 3. What is the difference between zeros, ones, eye, diagonal, unique, linspace functions, hstack and vstack functions, randint, rand, seed functions?
zeros: Creates an array filled with zeros. np.zeros((3, 3))
    
ones: Creates an array filled with ones. np.ones((3, 3))

eye: Creates an identity matrix. np.eye(3)

diagonal: Extracts or constructs a diagonal array. np.diag([1, 2, 3])

unique: Finds unique elements in an array. np.unique(array)


linspace: Generates linearly spaced values between two points. np.linspace(0, 10, 5)

hstack: Stacks arrays horizontally (column-wise). np.hstack((array1, array2))

vstack: Stacks arrays vertically (row-wise). np.vstack((array1, array2))

randint: Generates random integers. np.random.randint(0, 10, 5)

rand: Generates random floats in the range [0, 1). np.random.rand(3, 3)

seed: Sets the random seed for reproducibility. np.random.seed(42)

```python
a=np.zeros((3,3))
a
```

```python
a=np.ones((3,3))
a
```

```python
a=np.eye(3)
a
```

```python
a=np.diag([1,2,3])
a
```

```python
b=np.unique(a)
b
```

```python
c=np.linspace(0,10,5)
c
```

```python
c=np.hstack(a)
c
```

```python
c=np.vstack(a)
c
```

```python
c=np.random.randint(0, 10, 5)
c
```

```python
c= np.random.rand(3, 3)
c
```

```python
np.random.seed(42)
a=np.random.randint(1,100,10)
a
```

## 6. How do you handle missing data in Pandas?
Drop missing values: df.dropna()

Fill missing values: df.fillna(value)

Forward fill: df.fillna(method='ffill')

Backward fill: df.fillna(method='bfill')


```python
import pandas as pd
```

```python

# Sample DataFrame with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, np.nan, 35, 40],
    'Salary': [50000, 60000, np.nan, 80000],
    'Department': ['HR', 'IT', 'Finance', np.nan]
}

data_df = pd.DataFrame(data)
print("Original DataFrame:")
data_df
```

```python
dfx=data_df.dropna()
dfx
```

```python
dfx=data_df.fillna(0)
dfx
```

```python
dfx= data_df.fillna(method='ffill') # -- fills with the forward value
dfx
```

```python
dfx= data_df.fillna(method='bfill') # -- fills with the backward value
dfx
```

## 14. Difference between univariate, bivariate, and multivariate analysis with examples
Univariate: Analysis of a single variable. Example: Histogram of ages.

Bivariate: Analysis of two variables to find relationships. Example: Scatter plot of height vs. weight.

Multivariate: Analysis of more than two variables. Example: Multivariate regression.

## 15. Difference between discrete data and continuous data
Discrete Data: Countable and finite. Example: Number of students in a class.

Continuous Data: Measurable and infinite. Example: Height of students.
